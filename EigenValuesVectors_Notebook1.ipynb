# Eigen Values and Eigen Vectors

```python
#Import the necessary fuctions from the corresponding modules
from numpy import array,dot #To represent the matrix, vectors and multiplication of matrices
from numpy.linalg import eig # To find the Eigen Values
import matplotlib.pyplot as plt# To Represent the Eigen vectors visually
```

```python
matrix = np.array([[1,6],[5,2]]) #Representing the matrix as two dimenstional array
u= np.array([6,-5]) #Respresenting the eigen vectors u and v as a one dimensional array
v= np.array([3,-2])
```

The equation for eigenvalue is given by,
     Av = 位v
     where,
        A -> Matrix
        v -> Eigen Vector
        位 -> Eigen Value

```python
#From the eqution mentioned above find the LHS (mulitply matrix and eigen vector)
# Multiply matrix and u
Mu = dot(matrix,u)
#Multiply matrix and v
Mv = dot(matrix,v)
```

```python
#In RHS we have to find the 位 value
eigen_value,eigen_vector = eig(matrix)
print(eigen_value)
```

```python
#We have Av and Eigen Value , when we divide Av by v then we get the 位 value which we can check whether this values are Eigen values of Matrix
def check_eigenVector(mul,vec):
    div = mul/vec
    if(len(set(div))==1  and div[0] in eigen_value):
        print( vec," is a Eigen vector of the given matrix")
    else:
        print(vec," is not a Eigen vector of the given matrix")
check_eigenVector(Mu,u)
check_eigenVector(Mv,v)
```

```python
def plot_graph(vector,mul):
    origin = [0, 0] # unpacks the origin list [0, 0] into two arguments, representing the starting point of the vectors. 
    plt.quiver(*origin, vector[0], vector[1], angles='xy', scale_units='xy', scale=1, color='r',label=vector)
    plt.quiver(*origin, mul[0], mul[1], angles='xy', scale_units='xy', scale=1, color='b',label=mul)
    plt.xlim(-10, 10)
    plt.ylim(-10, 10)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title('Matrix * vector vs vector')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid()
    plt.show()
plot_graph(u,Mu)
plot_graph(v,Mv)
```

From the figure we can conclude that,
* first figure which is u is a Eigen vector Because the Mu and u in a same line 
* Second Figure which is v is not a Eigen vector because the Mv and v is not in the same line
