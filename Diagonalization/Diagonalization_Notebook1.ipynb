## Calculate D^2 to understand why diagonal matrices can be useful.

```python
import numpy as np #Import the necessary Modules
```

```python
D = np.array([[3,0],  #Define the Diagonal Matrix
             [0,6]]) 
print("Diagonal Matrix:D",D)
```

```python
def matrix_mul(mat1,mat2): # Matrix Multiplication which is required to calculate D^2
    if(len(mat1[0])!=len(mat2)):#Number of columns in mat1 must equal to number of number of rows in mat2 for matrix multiplication
        raise "Array Dimension does not match"
    result = np.zeros([len(mat1),len(mat2[0])])# Creating the result matrix which dimesion is equal to number of rows in first matrix , number of columns in second matrix
    for i in range(len(mat1)): #Iterate through rows in mat1
        for j in range(len(mat2)):#Iterate through columns in mat2
            for k in range(len(mat1[0])):#Iterate in the result matrix
                result[i][j]+=mat1[i][k]*mat2[k][j] #Perform multiplication
    return result
                
```

```python
print("Calcuation of D^2:",matrix_mul(D,D))
```

### Need of Diagonal Matrix:
    From the above result we can conclude that , when find the power of Digonal matrix , calculation is very simple,in which have to find powers of the diagonal elements alone

## Plot e1 and e2 vectors on a cooridnate system.Trasnform e1 and e2 using the Diagonal Matrix D. What do you notice?

```python
e1 = np.array([1,0]) #Define e1
e2 = np.array([0,1]) #Define e2
print("e1:",e1)
print("e2:",e2)
```

```python
import matplotlib.pyplot as plt #Import the necessary Modules.
```

```python
plt.quiver(0,0,e1[0],e1[1],scale_units="xy",angles='xy',color='r',label='e1',scale=1) # angle,lenghth of the line is based on the x and y axis.
plt.quiver(0,0,e2[0],e2[1],scale_units="xy",angles="xy",color='b',label='e2',scale=1)
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2,3)
plt.ylim(-1,3)
```

```python
def matrix_vector_mul(matrix,vector):
    if matrix.shape[1] != vector.shape[0]:
        raise ValueError("Matrix and vector dimensions are incompatible.")
    result = np.zeros(matrix.shape[0])  # Initialize the result vector.
    for i in range(matrix.shape[0]):  # Iterate though the rows of the matrix
        for j in range(matrix.shape[1]):  # Iterate over columns of the matrix 
            result[i] += matrix[i, j] * vector[j] #Mulitply each element in the matrix with the vector.
    return result
```

```python
trasnformed_e1 = matrix_vector_mul(D,e1)
trasnformed_e2 = matrix_vector_mul(D,e2)
print("Transformed e1:",e1)
print("Transformed e2:",e2)
```

```python
plt.quiver(0,0,e1[0],e1[1],scale_units="xy",angles='xy',color='r',label='e1',scale=1) #x=0, y=0, u=e1[0], v=e2[0]
plt.quiver(0,0,e2[0],e2[1],scale_units="xy",angles="xy",color='b',label='e2',scale=1)
plt.quiver(0,0,trasnformed_e1[0],trasnformed_e1[1],scale_units="xy",angles='xy',color='g',label='e1 trasnsform',scale=1)
plt.quiver(0,0,trasnformed_e2[0],trasnformed_e2[1],scale_units="xy",angles="xy",color='y',label='e2 trasnform',scale=1)
text_offset = 0.1
plt.text(e1[0] + text_offset, e1[1] + text_offset, 'e1', color='r') # Write the name of the vector above the arrow. 
plt.text(e2[0] + text_offset, e2[1] + text_offset, 'e2', color='b')
plt.text(trasnformed_e1[0] + text_offset, trasnformed_e1[1] + text_offset, 'e1 trasns', color='g')
plt.text(trasnformed_e2[0] + text_offset, trasnformed_e2[1] - text_offset, 'e2 trans', color='y')
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-2,7)
plt.ylim(-1,7)
```
