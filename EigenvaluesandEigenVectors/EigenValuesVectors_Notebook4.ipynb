## Step 1: Formulate the importance of each page as a linear algebra problem.

#### x1 = x3 + 1/2 x4
#### x2 = 1/3 x1
#### x3 = 1/3 x1 + 1/2 x2 + 1/2 x4
#### x4 = 1/3 x1 + 1/2 x2

## Step 2: Verify that the solution to the problem is an EigenVector.

### Step 2.1: Construct the matrix.

```python
import numpy as np # Import the necessary modules
```

```python
matrix = np.array([[0,0,1,1/2],
                   [1/3,0,0,0],
                   [1/3,1/2,0,1/2],
                   [1/3,1/2,0,0]])
```

### Step 2.2: Calcuate the Eigen Values and Eigen Vectors.

```python
eigValue,eigVector = np.linalg.eig(matrix)
print("EigenValues:",eigValue)
print("EigenVectors:",eigVector)
```

### Step 2.3: Extract the Eigen Vector that corresponding to the Eigen Value 1.

```python
for i in range(len(eigValue)):
    if np.isclose(eigValue[i], 1.0):#Comparisons made with small tolerance value.
        index = i
        break
solution = eigVector[:,index].real # Extracting the real values only.
print(solution)
```

## Step 3: Rescale the Eigen Vector as a probability by making sure sum of all entries in the eigenvector =1. 

```python
solution = solution / np.sum(solution)
print(solution)
```

## Step 4 : Display the Rank of each page.

```python
for i in range(len(solution)):
    print("Page:{} rank = {:.2f}".format(i+1,solution[i], 2))
```
