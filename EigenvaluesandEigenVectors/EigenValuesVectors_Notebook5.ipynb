```python
import numpy as np #Import the necessary Modules
```

```python
# Define the probability matrix p 
matrix = np.array([[0,0,1,0.5],
                   [0.333333,0,0,0],
                   [0.333333,0.5,0,0.5],
                   [0.333333,0.5,0,0]])
```

```python
def matrix_multiplication(matrix, x):
    if matrix.shape[1] != x.shape[0]: # For multiplication number of column in first matrix is equal to number of rows in second column
        raise Exception("Array Dimensions don't match")
    result = np.zeros((matrix.shape[0], x.shape[1])) #Creating an empty array to store the result of multiplication which dimestion is (number of rows in first matrix, number of column in second matrix) 
    for i in range(matrix.shape[0]): #first matrix's number of row time iteration
        for j in range(x.shape[1]): # second matirx's number of column time iteration
            for k in range(x.shape[0]): #first matrix's number of columsn time or second matrix number of row time 
                result[i][j] += matrix[i][k] * x[k][j] # multiply
    
    return result
```

## Question 1: Let’s say you start on Page 1, and then depending on this matrix, you move to the next page. Repeat this process, say a hundred times. How would the composition of vector x look ?

```python
# Starting from Page 1
x1 = matrix[:,0:1] #Making it as 2D array fo the matrix multiplication we can multiply the matrices if they are in same dimesnsions.
print("Page 1:",x1.flatten()) #printing it as 1D array for readability
for _ in range(100):
    x1 = matrix_multiplication(matrix, x1) #To move to the next state we have to multiply the current pages's probability with the matrix.(By markav's Chanin Rule)
print("Move to the next page 100 times.")
print("Composition of the matrix:")
print(x1.flatten())
```

## Question 2: Let’s say you start on Page 3, and then depending on this matrix, you move to the next page. Repeat this process, say a hundred times. How would the composition of vector x look ?

```python
# Starting from Page 3
x3 = matrix[:,2:3] #Making it as 2D array
print("Page 3:",x3.flatten())
for _ in range(100):
    x3 = matrix_multiplication(matrix, x3)
print("Move to the next page 100 times.")
print("Composition of the matrix:")    
print(x3.flatten())
```
